```python
import polars as pl
import holoviews as hv
```

```python
df = pl.read_csv("benchmark5.csv").with_columns(pl.concat_str([pl.col("name"), pl.col("operation"), pl.col("key size(B)"), pl.col("value size(B)"), pl.col("thread")], separator="/").alias("label"))
```

```python
df_set = df.filter(pl.col("operation") == "set")
df_get = df.filter(pl.col("operation") == "get")
```

```python
def grid_plot(df, x, y, cols=2, thread=16):
    plots = []
    for k in [2 ** 2, 2 ** 4, 2 ** 6, 2 ** 8, 2 ** 10, 2 ** 12, 2 ** 14]:
        for v in [1024]:
            plots.append((df
                .filter(pl.col("key size(B)") == k)
                .filter(pl.col("value size(B)") == v)
                .filter(pl.col("thread") == thread)
                .plot.scatter(x=x, y=y, by="label")
            ))
    grid = hv.Layout(plots).cols(cols)
    grid.opts(hv.opts.Scatter(tools=['hover']))
    return grid
```

```python
grid_plot(df_get, "time(ns)", "operation count")
```

```python
grid_plot(df_set, "time(ns)", "operation count")
```

```python
grid_plot(df_set, "operation count", "vmrss(KB)")
```

```python
grid_plot(df_set, "operation count", "file size(B)")
```
